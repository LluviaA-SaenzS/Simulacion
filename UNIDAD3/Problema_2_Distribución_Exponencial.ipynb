**Distribución Exponencial**

Una estación de carga rápida para vehículos eléctricos quiere
simular el tiempo entre la llegada de los vehículos para prever la cantidad
de puntos de carga necesarios durante el día. La llegada de los vehículos
sigue una distribución exponencial con una tasa de 1 llegada cada 15
minutos, en promedio.

* **Objetivo:** Generar tiempos entre llegadas de vehículos eléctricos durante
un día de operación (12 horas).
---


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lambda_llegada = 1 / 15  # Tiempo promedio entre llegadas = 15 minutos

# Función para generar tiempo entre llegadas
def generar_tiempo_entre_llegadas(lambda_llegada):
    u = np.random.uniform(0, 1)  # número aleatorio uniforme
    tiempo_entre_llegadas = -1 / lambda_llegada * np.log(1 - u)
    return tiempo_entre_llegadas

# Generar tiempos entre llegadas hasta completar 12 horas de operación (720 minutos)
duracion_dia = 12 * 60  # Duración en minutos
tiempos_entre_llegadas = []
tiempo_total = 0

# Simular tiempos entre llegadas acumulando hasta 12 horas
while tiempo_total < duracion_dia:
    tiempo = generar_tiempo_entre_llegadas(lambda_llegada)
    tiempos_entre_llegadas.append(tiempo)
    tiempo_total += tiempo
```

```python

print("Tiempos entre llegadas generados:", tiempos_entre_llegadas)
plt.hist(tiempos_entre_llegadas, bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribución de Tiempos entre Llegadas Simulados')
plt.xlabel('Tiempo entre llegadas (minutos)')
plt.ylabel('Frecuencia')
plt.show()
```

**Tiempo entre llegadas:** Representa el tiempo entre la llegada de un vehículo y el siguiente, en minutos.

**Frecuencia:** Muestra cuántas veces aparece cada intervalo de tiempo entre llegadas en el conjunto de datos simulados para las 12 horas de operación.

# **CONCLUSION**
La mayoría de los intervalos entre llegadas están cerca de los 15 minutos, como se esperaba.

Existen algunos intervalos más largos entre llegadas, lo que indica que aunque la mayoría de las llegadas ocurren con intervalos cortos, ocasionalmente se observan períodos de menor actividad en la estación.

Esta simulación puede ayudar a la estación a preveer cuántos puntos de carga serán necesarios en momentos de alta frecuencia de llegada de vehículos y cómo ajustar el número de puntos en función de la demanda.
