```python
import numpy as np
import matplotlib.pyplot as plt

lambda_servicio =1/2

def generar_tiempo(lambda_servicio):
  u= np.random.uniform(0,1)
  tiempo_servicio= -1/lambda_servicio * np.log(1 - u)
  return tiempo_servicio

tiempo_servicio = [generar_tiempo(lambda_servicio) for _ in range(100)]
print('Tiempo de servicio: ', tiempo_servicio)
plt.hist(tiempo_servicio,bins=10,edgecolor='black',alpha=0.7)
plt.title('Distribucion de tiempos de servicio simulado')
plt.xlabel('Tiempo minutos')
plt.ylabel('frecuencia')
```
